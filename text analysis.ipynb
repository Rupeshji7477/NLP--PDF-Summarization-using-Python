```python
# import gensim
import nltk 
import PyPDF2
import re 
import heapq
import fitz

file_Path = 'Operations Management.pdf'

# with open(file_Path, mode = 'rb') as file:
  
#     pdf_reader = PyPDF2.PdfReader(file)
 # for page_number in range(len(pdf_reader.pages)):
    #     page = pdf_reader.pages[page_number]
    #     text += page.extract_text()
with fitz.open(file_Path)as pdf:    
   
    text = ""
    for page in pdf:
        text += page.get_text()
   
print(text)
```

```python
# # removing square brackets and extra spaces
text = re.sub(r'\[[0-9]*\]', ' ',text)
text = re.sub(r'\s+',' ',text)
print (text )
```

```python
# # removing  special character and digits
Clean_text = re.sub('[^a-zA-Z]',' ',text)
Clean_text = re.sub(r'\s+',' ',Clean_text)
print (Clean_text)
```

```python
sentence_list = nltk.sent_tokenize(text)
 
```

```python
# wieghted frequency of Occurrence
stopwords = nltk.corpus.stopwords.words('english')

word_frequencies = {}
for word in nltk.word_tokenize(Clean_text):
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word]=1
        else:
            word_frequencies[word] += 1
 
```

```python
#calculated weighted frequency of each word
maximum_frequency = max(word_frequencies.values())

for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequency)
```

```python
# calculating Sentence Scores
 
sentence_scores = {}
for sent in sentence_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' '))<30:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
                    

```

```python
summary_sentence = heapq.nlargest(22, sentence_scores , key = sentence_scores.get)

summary = ' '.join(summary_sentence)
summary_with_line_break = summary.replace('. ','.\n')
print(summary)
word_count = len(summary.split())
print(word_count)
```

```python
with open ('summary.pdf','w') as file:
    file.write(summary)
```
